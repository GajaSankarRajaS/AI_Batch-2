## Part 1 — History of AI (≈25 min)

### 1.1 The Birth of an Idea (1950s)

> *"Can machines think?"* — Alan Turing, 1950

Turing proposed the **Turing Test**: if a human can't reliably tell whether
they're talking to a machine or a person, the machine can be said to "think."

The term **"Artificial Intelligence"** itself was coined in **1956** at the
Dartmouth Conference, where researchers optimistically predicted machines
would match human intelligence within a generation. They were... a little early.


### 1.2 The Symbolic Era (1950s–1980s)

Early AI was built on **logic and rules** — the idea that intelligence is
fundamentally about *manipulating symbols* according to formal rules.

- **Expert systems**: programs that encoded human expert knowledge as
  "if-then" rules (e.g., medical diagnosis systems)
- **Strengths**: explainable, precise for narrow domains
- **Weaknesses**: brittle — couldn't handle anything outside their rules,
  and writing rules for "common sense" proved nearly impossible

This led to the famous **"AI Winters"** — periods (mid-1970s and late-1980s)
where funding and interest collapsed because progress stalled.


### 1.3 The Machine Learning Era (1990s–2010s)

Instead of hand-writing rules, what if machines could **learn patterns from data**?

- Statistical methods, decision trees, support vector machines
- **1997**: IBM's Deep Blue beats world chess champion Garry Kasparov
- **2000s**: Rise of "Big Data" — more data + more compute = better models
- **2012**: A turning point — a neural network (AlexNet) dramatically
  outperforms older methods on image recognition, kickstarting the
  **deep learning revolution**


### 1.4 The Modern Era (2017–Present)

- **2017**: The **Transformer** architecture is introduced — the foundation
  of nearly all modern language AI
- **2020s**: Large Language Models (GPT, Claude, Gemini) bring AI into
  everyday life — writing, coding, image generation, voice assistants
- AI shifts from "a research topic" to "a tool in every industry"

## Part 2 — Narrow AI vs. General AI vs. Super AI (≈15 min)

AI systems are often classified by their **scope of intelligence**:


### 2.1 Narrow AI (ANI — Artificial Narrow Intelligence)

- Designed to perform **one specific task** very well
- This is **everything that exists today** — including ChatGPT, Claude,
  self-driving cars, recommendation systems, spam filters
- A chess AI can't suddenly decide to write poetry — it does one thing

### 2.2 General AI (AGI — Artificial General Intelligence)

- A hypothetical system that can understand, learn, and apply
  intelligence across **any task a human can**, with human-level flexibility
- Does **not exist yet** — it's a major research goal and topic of debate
  about timelines (some say decades, some say never, some say soon)

### 2.3 Super AI (ASI — Artificial Super Intelligence)

- A hypothetical AI that **surpasses human intelligence** across all domains
- Purely theoretical — discussed mostly in the context of AI safety and
  long-term risk research

## Part 3 — Key Subfields of AI (≈25 min)

AI is an umbrella term. Here are the major subfields you'll encounter
throughout this course (and in your career):


### 3.1 Machine Learning (ML)

The study of algorithms that **improve through experience (data)** rather
than explicit programming.

- **Supervised learning**: learn from labeled examples (e.g., spam vs. not spam)
- **Unsupervised learning**: find patterns in unlabeled data (e.g., clustering customers)
- **Reinforcement learning**: learn by trial-and-error with rewards (e.g., game-playing agents)

*Real-world example: Netflix recommendations, credit scoring*


### 3.2 Natural Language Processing (NLP)

Enables machines to understand, interpret, and generate **human language**.

- Translation, sentiment analysis, chatbots, summarization
- Modern NLP is powered by **Large Language Models** (which we cover Day 13–15)

*Real-world example: Google Translate, voice assistants, Claude/ChatGPT*


### 3.3 Computer Vision (CV)

Enables machines to **interpret visual information** — images and video.

- Object detection, facial recognition, medical image analysis
- Self-driving cars rely heavily on CV to "see" the road

*Real-world example: Face unlock on phones, Google Lens, quality inspection in factories*


### 3.4 Robotics

Combines AI with physical systems that **sense and act in the real world**.

- Requires perception (CV/sensors), reasoning (planning), and action (motors/actuators)
- Often the hardest subfield because the real world is messy and unpredictable

*Real-world example: warehouse robots, robotic arms in manufacturing, drones*

## Part 4 — How AI Systems Are Structured (≈10 min)

Almost every AI system — no matter how complex — follows the same basic loop:

```
   PERCEPTION  →  REASONING  →  ACTION
   (sensors/    (processing/   (output/
    input data)  decision-      effect on
                  making)        environment)
        ↑                              |
        └──────── feedback ────────────┘
```

- **Perception**: How does the system receive information? (camera, text input, sensor data, microphone)
- **Reasoning**: What does it *do* with that information? (a trained model, a search algorithm, a set of rules)
- **Action**: What does it output or do? (a prediction, a recommendation, a physical movement)
- **Feedback**: Many systems learn from the results of their actions, closing the loop

This is exactly the framework we'll formalize on **Day 6 (Intelligent Agents)** —
today is about building intuition; Day 6 gives it precise vocabulary.

## Part 5 — Current State & Future Outlook (≈10 min)

### Where AI stands today (2026)

- AI assistants are integrated into everyday workflows (writing, coding, research)
- Multimodal models can process text, images, audio, and video together
- AI is being deployed in healthcare diagnostics, scientific research, and
  creative industries — alongside ongoing debate about jobs, ethics, and regulation
- We'll dig deeper into the **ethical and societal questions** on Days 11–12

### Where it might be heading

- More **agentic** systems — AI that can take multi-step actions on your behalf
  (you've already glimpsed this with the Wumpus World agent on Day 5!)
- Continued debate about AGI timelines and AI safety
- Growing focus on making AI systems more **efficient, transparent, and trustworthy**
