## 2️⃣ Variables

A **variable** is a named box that stores a value in the computer's memory, so you can use it later.

```
   name        =        value
 ┌────────┐          ┌─────────┐
 │  age   │  <-----   │   20    │
 └────────┘  stores   └─────────┘
```

Syntax: `variable_name = value`

The `=` here is **not** "equals" like in math — it means "assign this value to this name".

````python
age = 20
name = "Gaja"
height_in_cm = 175.5

print(name)
print(age)
print(height_in_cm)
````

### Rules for naming variables
- Must start with a letter or underscore (`_`), not a number
- Can contain letters, numbers, underscores — no spaces
- Case-sensitive (`age` and `Age` are different variables)
- Avoid Python keywords like `print`, `if`, `for`

```
✅ good_name       ✅ _score       ✅ score2
❌ 2score          ❌ my name      ❌ class
```

````python
# valid
student_name = "Ravi"
_temp = 98.6
score2 = 88

print(student_name, _temp, score2)
````

### Reassigning variables
A variable's value can change — that's why it's called a *variable*.

````python
balance = 100
print("Before:", balance)

balance = 250
print("After:", balance)
````

### Data types of variables
Every value has a **type**. The most common beginner types:

| Type | Example | Meaning |
|---|---|---|
| `int` | `10` | whole number |
| `float` | `10.5` | decimal number |
| `str` | `"hello"` | text |
| `bool` | `True` / `False` | true or false |

Use `type()` to check a variable's type.

````python
x = 10
y = 3.14
name = "Python"
is_fun = True

print(type(x))
print(type(y))
print(type(name))
print(type(is_fun))
````

## 3️⃣ Operators

**Operators** are special symbols that perform operations on values (called **operands**).

```
   5     +     3     =     8
   ↑     ↑     ↑           ↑
operand operator operand  result
```

Example: in `5 + 3`, `+` is the operator, and `5`, `3` are operands.

````python
a = 10
b = 3

print(a + b)   # addition
````

## 4️⃣ Types of Operators

Python has several categories of operators. We'll cover the 5 most important for beginners:

```
┌─────────────────────────────────────────┐
│              OPERATORS                  │
├───────────────┬───────────────────────  │
│ Arithmetic     │  +  -  *  /  //  %  ** │
│ Comparison     │  == != > < >= <=       │
│ Logical        │  and  or  not          │
│ Assignment     │  =  +=  -=  *=  /=     │
│ Membership     │  in  not in            │
└───────────────┴───────────────────────  │
```


### 4.1 Arithmetic Operators
Used for math.

| Operator | Meaning | Example | Result |
|---|---|---|---|
| `+` | addition | `7 + 2` | `9` |
| `-` | subtraction | `7 - 2` | `5` |
| `*` | multiplication | `7 * 2` | `14` |
| `/` | division (always float) | `7 / 2` | `3.5` |
| `//` | floor division (rounds down) | `7 // 2` | `3` |
| `%` | modulus (remainder) | `7 % 2` | `1` |
| `**` | exponent (power) | `7 ** 2` | `49` |

````python
a = 7
b = 2

print("a + b  =", a + b)
print("a - b  =", a - b)
print("a * b  =", a * b)
print("a / b  =", a / b)
print("a // b =", a // b)
print("a % b  =", a % b)
print("a ** b =", a ** b)
````

### 4.2 Comparison Operators
Compare two values → always give back `True` or `False`.

| Operator | Meaning |
|---|---|
| `==` | equal to |
| `!=` | not equal to |
| `>` | greater than |
| `<` | less than |
| `>=` | greater than or equal to |
| `<=` | less than or equal to |

⚠️ Common beginner mistake: `=` assigns a value, `==` compares values.

````python
x = 10
y = 20

print(x == y)   # False
print(x != y)   # True
print(x > y)    # False
print(x < y)    # True
print(x >= 10)  # True
````

### 4.3 Logical Operators
Combine multiple `True`/`False` conditions.

| Operator | Meaning |
|---|---|
| `and` | True only if BOTH sides are True |
| `or`  | True if AT LEAST ONE side is True |
| `not` | flips True ↔ False |

````python
age = 22
has_id = True

print(age >= 18 and has_id)   # both true -> True
print(age >= 18 or has_id)    # at least one true -> True
print(not has_id)             # flips True -> False
````

### 4.4 Assignment Operators
Shortcuts to update a variable's value.

| Operator | Same as |
|---|---|
| `x += 5` | `x = x + 5` |
| `x -= 5` | `x = x - 5` |
| `x *= 5` | `x = x * 5` |
| `x /= 5` | `x = x / 5` |

````python
score = 10
print("Start:", score)

score += 5   # score = score + 5
print("After +=5:", score)

score *= 2   # score = score * 2
print("After *=2:", score)
````

### 4.5 Membership Operators
Check if a value exists inside a collection (like a string or list).

| Operator | Meaning |
|---|---|
| `in` | True if value is found |
| `not in` | True if value is NOT found |

````python
fruits = ["apple", "banana", "mango"]

print("banana" in fruits)      # True
print("grapes" in fruits)      # False
print("grapes" not in fruits)  # True

# works on strings too
print("a" in "banana")
````
