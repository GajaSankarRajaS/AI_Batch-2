## 1️⃣ Dictionaries

A **dictionary** stores data as **key-value pairs**. Think of a real dictionary: you look up a word (key) to get its definition (value).

```
        Dictionary: student
    ┌──────────────────────────┐
    │  name → "Alice"          │
    │  age → 20                │
    │  city → "Mumbai"         │
    └──────────────────────────┘
    (key → value pairs)
```

**Syntax:** `{key1: value1, key2: value2, ...}`

Key traits:
- **Unordered** (but in Python 3.7+, order is preserved)
- **Mutable** — can add/remove/change pairs
- **Keys must be unique** — one key per dictionary
- **Keys must be immutable** — typically strings or numbers
- **Values can be anything** — strings, numbers, lists, even other dicts


````python
student = {
    "name": "Alice",
    "age": 20,
    "city": "Mumbai",
    "gpa": 3.8
}

print(student)
print(type(student))
````

### Accessing values
Use the key inside square brackets `[key]` or `.get(key)`.

````python
student = {
    "name": "Alice",
    "age": 20,
    "city": "Mumbai"
}

print(student["name"])    # access by key
print(student["age"])
print(student["city"])
````

### `.get()` method
Safer way to access — returns `None` if key doesn't exist (no error).

````python
student = {"name": "Alice", "age": 20}

print(student.get("name"))      # "Alice"
print(student.get("city"))      # None (key doesn't exist)
print(student.get("city", "N/A"))  # with default value
````

### Modifying dictionaries
Add, update, or delete key-value pairs.

````python
student = {"name": "Alice", "age": 20}
print("Before:", student)

# Add/update
student["city"] = "Mumbai"     # add new key
student["age"] = 21            # update existing key
print("After update:", student)

# Delete
del student["city"]
print("After deletion:", student)
````

## 2️⃣ Dictionary Methods

### 2.1 Methods to access data

#### `.keys()` — Get ALL keys


````python
student = {"name": "Alice", "age": 20, "city": "Mumbai"}

keys = student.keys()
print("Keys:", keys)
print("Type:", type(keys))

# Convert to list if needed
keys_list = list(keys)
print("As list:", keys_list)
````

#### `.values()` — Get ALL values

````python
student = {"name": "Alice", "age": 20, "city": "Mumbai"}

values = student.values()
print("Values:", values)

# Convert to list
values_list = list(values)
print("As list:", values_list)
````

#### `.items()` — Get ALL key-value pairs

````python
student = {"name": "Alice", "age": 20, "city": "Mumbai"}

items = student.items()
print("Items:", items)

# Useful in loops
for key, value in student.items():
    print(f"{key}: {value}")
````

### 2.2 Methods to modify

#### `.update(other_dict)` — Merge another dictionary


````python
student = {"name": "Alice", "age": 20}
print("Before:", student)

more_info = {"city": "Mumbai", "gpa": 3.8}
student.update(more_info)
print("After update:", student)
````

#### `.pop(key)` — Remove a key-value pair and return the value

````python
student = {"name": "Alice", "age": 20, "city": "Mumbai"}
print("Before:", student)

removed = student.pop("city")
print("Removed value:", removed)
print("After:", student)
````

#### `.clear()` — Remove ALL key-value pairs

````python
student = {"name": "Alice", "age": 20}
print("Before:", student)

student.clear()
print("After:", student)
````

### 2.3 Methods to check/copy

#### `.copy()` — Make an independent copy


````python
original = {"name": "Alice", "age": 20}
copy_dict = original.copy()

copy_dict["age"] = 21
copy_dict["city"] = "Mumbai"

print("Original:", original)
print("Copy:", copy_dict)
````

#### `.setdefault(key, default)` — Get value or set default if key missing


````python
student = {"name": "Alice", "age": 20}

age = student.setdefault("age", 18)
print("Age:", age)

city = student.setdefault("city", "Unknown")
print("City:", city)
print("Dictionary after:", student)
````

### 2.4 Quick reference table

| Method | What it does | Example |
|--------|-------------|---------|
| keys() | Get all keys | d.keys() |
| values() | Get all values | d.values() |
| items() | Get all key-value pairs | d.items() |
| get(key) | Get value (safe) | v = d.get("name") |
| update(d2) | Merge dictionaries | d.update(d2) |
| pop(key) | Remove and return | v = d.pop("age") |
| clear() | Delete all | d.clear() |
| copy() | Make copy | d2 = d.copy() |
| setdefault(k, default) | Get or set | d.setdefault("age", 18) |

## 3️⃣ Tuples

A **tuple** is like a **locked list** — an ordered collection that **cannot be changed** after creation.

```
List:  [1, 2, 3]  <- can change
Tuple: (1, 2, 3)  <- locked, immutable
```

**Syntax:** `(item1, item2, item3, ...)` or just `item1, item2, item3`

Key traits:
- **Ordered** — position matters
- **Immutable** — CANNOT add/remove/change items
- **Can contain duplicates**
- **Faster than lists** (small performance benefit)
- **Can be dictionary keys** (lists cannot)


````python
coordinates = (10, 20, 30)
print(coordinates)
print(type(coordinates))

# Without parentheses (still a tuple)
point = 5, 15
print(point)
````

### Accessing tuples (same as lists)
Use indexing and slicing.

````python
colors = ("red", "green", "blue")

print(colors[0])       # first item
print(colors[-1])      # last item
print(colors[1:])      # from index 1 to end
print(colors[::2])     # every 2nd item
````

### Single-item tuple
Need a comma to distinguish from parentheses!

````python
single = (5,)      # tuple
not_tuple = (5)        # just a number in parentheses

print(type(single))
print(type(not_tuple))
````

### Creating tuples
Different ways to create tuples:

````python
t1 = (1, 2, 3)
t2 = tuple([1, 2, 3])        # from list
t3 = tuple("abc")            # from string
t4 = 1, 2, 3                 # no parentheses needed

print("t1:", t1)
print("t2:", t2)
print("t3:", t3)
print("t4:", t4)
````

### Immutability example
Tuples cannot be changed after creation.

````python
tpl = (1, 2, 3)

# Try to change (this would error):
# tpl[0] = 99

# But you CAN create a new tuple
tpl = (1, 2, 3)
tpl = (99,) + tpl[1:]    # create new tuple
print("New tuple:", tpl)
````

## 4️⃣ Tuple Methods

Tuples have only 2 methods (since they're immutable, can't add/remove):

### `.count(item)` — Count occurrences of an item


````python
numbers = (1, 2, 2, 3, 3, 3, 4)

count_3 = numbers.count(3)
count_1 = numbers.count(1)

print(f"3 appears {count_3} times")
print(f"1 appears {count_1} times")
````

### `.index(item)` — Find position of first occurrence

````python
colors = ("red", "green", "blue", "green")

idx = colors.index("green")
print(f"First 'green' is at index {idx}")
````

### Tuple unpacking
Extract items into separate variables.

````python
point = (10, 20)
x, y = point
print(f"x = {x}, y = {y}")

# Works with any number of items
rgb = (255, 128, 0)
r, g, b = rgb
print(f"RGB: ({r}, {g}, {b})")

# Skip items with _
coords = (10, 20, 30, 40)
x, y, _, _ = coords
print(f"First two: {x}, {y}")
````

### Tuple methods quick reference

| Method | What it does | Example |
|--------|-------------|---------|
| count(x) | Count x | t.count(3) |
| index(x) | Find position of x | i = t.index(3) |

**Remember:** Tuples are immutable! No add/remove/change methods.

## 5️⃣ Strings

A **string** is a sequence of **characters** (text). Strings are **immutable** like tuples.

**Syntax:** `"text"` or `'text'` or triple quotes

Key traits:
- **Ordered** — characters have positions
- **Immutable** — cannot change individual characters
- **Indexable & sliceable** — like lists and tuples
- **Iterable** — can loop through characters


````python
text = "Hello, Python!"
print(text)
print(type(text))

# Single vs double quotes (same thing)
s1 = "Hello"
s2 = 'Hello'
print(s1 == s2)
````

### Accessing characters
Use indexing and slicing.

````python
text = "Hello"

print(text[0])       # first character
print(text[-1])      # last character
print(text[1:3])     # slice
print(text[::-1])    # reverse
````

### Concatenation and repetition

````python
s1 = "Hello"
s2 = "World"

print(s1 + " " + s2)      # concatenation
print(s1 * 3)             # repetition
print("-" * 10)
````

### Escape sequences
Special characters using backslash.

````python
print("Line 1\nLine 2")      # newline
print("Tab\tseparated")       # tab
print("Backslash: \\")          # backslash
````

## 6️⃣ String Methods

Strings have TONS of built-in methods. Here are the most useful:

### 6.1 Case methods

#### `.upper()` — Convert to UPPERCASE


````python
text = "Hello, World!"
print(text.upper())
````

#### `.lower()` — Convert to lowercase

````python
text = "Hello, World!"
print(text.lower())
````

#### `.capitalize()` — First letter uppercase, rest lowercase

````python
text = "hello, world!"
print(text.capitalize())
````

#### `.title()` — Capitalize first letter of each WORD

````python
text = "hello world python"
print(text.title())
````

#### `.swapcase()` — Swap uppercase and lowercase

````python
text = "Hello, World!"
print(text.swapcase())
````

### 6.2 Search methods

#### `.find(substring)` — Find position of substring


````python
text = "Hello, Python!"

pos = text.find("Python")
print(f"Position: {pos}")

pos2 = text.find("Java")
print(f"Position of 'Java': {pos2}")  # -1 (not found)
````

#### `.index(substring)` — Same as find, but ERROR if not found

````python
text = "Hello, Python!"

pos = text.index("Python")
print(f"Position: {pos}")
````

#### `.count(substring)` — Count occurrences

````python
text = "Mississippi"
count = text.count("s")
print(f"'s' appears {count} times")

# Count substrings
text2 = "apple apple apple"
count2 = text2.count("apple")
print(f"'apple' appears {count2} times")
````

#### `.startswith()` and `.endswith()` — Check prefix and suffix

````python
url = "https://example.com"

print(url.startswith("https"))   # True
print(url.endswith(".com"))       # True
print(url.startswith("http"))    # False
````

### 6.3 Modification methods

#### `.replace(old, new)` — Replace substring


````python
text = "Hello, World!"
print("Original:", text)

new_text = text.replace("World", "Python")
print("After replace:", new_text)
print("Original unchanged:", text)  # immutable!

# Replace multiple occurrences
text2 = "apple apple apple"
new_text2 = text2.replace("apple", "orange")
print(new_text2)
````

#### `.strip()` — Remove whitespace from both ends

````python
text = "  hello  "
print(f"Original: '{text}'")
print(f"After strip: '{text.strip()}'")

# lstrip() and rstrip() for left/right only
print(f"Left strip: '{text.lstrip()}'")
print(f"Right strip: '{text.rstrip()}'")
````

#### `.split(delimiter)` — Split into LIST of substrings

````python
text = "apple,banana,mango"
fruits = text.split(",")
print(fruits)

# Split by space (default)
sentence = "Hello World Python"
words = sentence.split()
print(words)
````

#### `.join(list)` — Join list into STRING

````python
fruits = ["apple", "banana", "mango"]
text = ", ".join(fruits)
print(text)

# Works with string items
words = ["Hello", "World", "Python"]
sentence = " ".join(words)
print(sentence)
````

### 6.4 Check methods

#### `.isdigit()`, `.isalpha()`, `.isalnum()`, `.isspace()`


````python
print("123".isdigit())       # True
print("abc".isalpha())       # True
print("abc123".isalnum())    # True
print("   ".isspace())       # True
print("abc123".isalpha())    # False (contains numbers)
````

### 6.5 F-strings (formatted strings)
Modern, readable way to insert variables into strings.

````python
name = "Alice"
age = 20
gpa = 3.8

# F-string (modern way)
msg = f"Name: {name}, Age: {age}, GPA: {gpa:.1f}"
print(msg)

# Expressions inside braces
print(f"Age next year: {age + 1}")
print(f"Name in caps: {name.upper()}")
````

| join(list) | Join to string | ",".join(list) |

| isalpha() | All letters | s.isalpha() |

## 7️⃣ Data Types Comparison

| Feature | List | Tuple | Dict | String |
|---------|------|-------|------|--------|
| Ordered | Yes | Yes | Yes (3.7+) | Yes |
| Indexed | Yes | Yes | No | Yes |
| Mutable | Yes | No | Yes | No |
| Duplicates | Yes | Yes | No (keys) | Yes |
| Use case | Sequence | Immutable sequence | Key-value pairs | Text |

### When to use what?

**Lists:** `[1, 2, 3]` - need to modify, order matters

**Tuples:** `(1, 2, 3)` - immutable, hashable (dict keys), slight speed boost

**Dicts:** `{"name": "Alice"}` - key-value relationships, fast lookup

**Strings:** `"hello"` - text, immutable, tons of methods


---
## Summary

**Dictionaries:**
- Key-value pairs: `{"key": value}`
- Methods: .keys(), .values(), .items(), .get(), .pop(), .update()

**Tuples:**
- Immutable sequences: `(1, 2, 3)`
- Methods: .count(), .index() only
- Unpacking: `x, y = (1, 2)`

**Strings:**
- Text sequences: `"hello"`
- Methods: .upper(), .lower(), .find(), .replace(), .split(), .join()
- F-strings: `f"text {variable}"`
